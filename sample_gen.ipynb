# じゃらんデータ整備のためにサンプルデータを作る

## データの読み込み

```python
import csv
# csv_file出力
import os
import pandas as pd
# Shift-JIS 形式の CSV ファイル (employee.sjis.csv) して出力
# df.to_csv("employee.sjis.csv", encoding="shift_jis")
# csvを開く。

os.chdir("D:\じゃらん宿泊データ\original_data")

csv_file_name = "hotel_res2021-04-01.csv"
columns = ["エリアコード","宿コード","データ取得日","チェックイン日","プランコード","価格","部屋情報","朝食の有無","夕食の有無","チェックイン時刻","チェックアウト時刻","プラン提供開始日","プラン提供終了日","イメージ画像URL","画像キャプション","残り部屋数"]

df = pd.read_csv(csv_file_name,names=(columns))
df.head()
```

## データを10000行だけ抽出

```python
df_sample = df.iloc[:10000]
```

## テキストをまとめて元のデータを削除

```python
text = df_sample["部屋情報"] + df_sample["画像キャプション"]
df_s_text = pd.DataFrame(text,columns=["text"])
df_s_text
```

```python
df_s = df_sample.drop(columns=["部屋情報","イメージ画像URL","画像キャプション"])
df_s = pd.concat([df_s, df_s_text], axis=1)
df_s
```

## フラグの追加

```python
kinnen = []
kitsuen = []
for text in df_s["text"]:
    text = str(text)
    kinnen.append("禁煙" in text)
    kitsuen.append("喫煙" in text)
#print(kinnen)
#print(kituen)
df_kinnen = pd.DataFrame(kinnen,columns=["禁煙"])
df_kitsuen = pd.DataFrame(kitsuen,columns=["喫煙"])
df_s = pd.concat([df_s,df_kinnen,df_kitsuen], axis=1)
df_s = df_s.drop(columns="text")
df_s
```

## サンプルデータをcsvファイルで保存

```python
os.chdir("D:\じゃらん宿泊データ\sample_data")
df_s.to_csv(csv_file_name)
```
