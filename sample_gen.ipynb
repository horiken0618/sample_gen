# じゃらんデータ整備のためにサンプルデータを作る

## データの読み込み

```python
import csv
# csv_file出力
import os
import pandas as pd
# Shift-JIS 形式の CSV ファイル (employee.sjis.csv) して出力
# df.to_csv("employee.sjis.csv", encoding="shift_jis")
# csvを開く。

os.chdir("D:\じゃらん宿泊データ\original_data")

csv_file_name = "hotel_res2021-04-01.csv"
columns = ["エリアコード","宿コード","データ取得日","チェックイン日","プランコード","価格","部屋情報","朝食の有無","夕食の有無","チェックイン時刻","チェックアウト時刻","プラン提供開始日","プラン提供終了日","イメージ画像URL","画像キャプション","残り部屋数"]

df = pd.read_csv(csv_file_name,names=(columns))
df.head()
```

## データを10000行だけ抽出

```python
# df_sample = df.iloc[:10000]
df_sample = df.copy()
```

## テキストをまとめて元のデータを削除

```python
text = df_sample["部屋情報"] + df_sample["画像キャプション"]
df_s_text = pd.DataFrame(text,columns=["text"])
df_s_text
```

```python
df_s = df_sample.drop(columns=["部屋情報","イメージ画像URL","画像キャプション"])
df_s = pd.concat([df_s, df_s_text], axis=1)
df_s
```

```python
len(df_s)
```

## フラグの追加

```python
kinnen = []
kitsuen = []
wa = []
yo = []
wayo = []
single = []
semidouble = []
double = []
twin = []
tri = []
omakase = []
renpaku = []

for text in df_s["text"]:
    text = str(text)
    kinnen.append(int("禁煙" in text))
    kitsuen.append(int("喫煙" in text))
    wa.append(int("和室" in text))
    yo_wa = int("和洋室" in text) + int("洋室" in text)
    if yo_wa == 2:
        wayo.append(1)
        yo.append(0)
    elif yo_wa == 1:
        wayo.append(0)
        yo.append(1)
    else:
        wayo.append(0)
        yo.append(0)
    single.append(int("シングル" in text))
    semi_dou = int("セミダブル" in text) + int("ダブル" in text)
    if semi_dou == 2:
        semidouble.append(1)
        double.append(0)
    elif semi_dou == 1:
        semidouble.append(0)
        double.append(1)
    else:
        semidouble.append(0)
        double.append(0)        
    twin.append(int("ツイン" in text))
    tri.append(int("トリプル" in text))
    omakase.append(int("おまかせ" in text))
    renpaku.append(int("連泊" in text))

#print(kinnen)
#print(kituen)
df_kinnen = pd.DataFrame(kinnen,columns=["禁煙"])
df_kitsuen = pd.DataFrame(kitsuen,columns=["喫煙"])
df_wa =  pd.DataFrame(wa,columns=["和室"])
df_yo =  pd.DataFrame(yo,columns=["洋室"])
df_wayo =  pd.DataFrame(wayo,columns=["和洋室"])
df_single =  pd.DataFrame(single,columns=["シングル"])
df_semidouble =  pd.DataFrame(semidouble,columns=["セミダブル"])
df_double =  pd.DataFrame(double,columns=["ダブル"])
df_twin =  pd.DataFrame(twin,columns=["ツイン"])
df_tri =  pd.DataFrame(tri,columns=["トリプル"])
df_omakase =  pd.DataFrame(omakase,columns=["おまかせ"])
df_renpaku = pd.DataFrame(renpaku,columns=["連泊"])
df_s = pd.concat([df_s,df_kinnen,df_kitsuen,df_wa,df_yo,df_wayo,df_single,df_semidouble,df_double,df_twin,df_tri,df_omakase,df_renpaku], axis=1)
df_s = df_s.drop(columns="text")
df_s
```

```python
# こっちの方が20%速い
import numpy as np

kinnen = []
kitsuen = []
wa = []
yo = []
wayo = []
single = []
semidouble = []
double = []
twin = []
tri = []
omakase = []
renpaku = []

for text in df_s["text"]:
    text = str(text)
    kinnen.append(int("禁煙" in text))
    kitsuen.append(int("喫煙" in text))
    wa.append(int("和室" in text))
    yo_wa = int("和洋室" in text) + int("洋室" in text)
    if yo_wa == 2:
        wayo.append(1)
        yo.append(0)
    elif yo_wa == 1:
        wayo.append(0)
        yo.append(1)
    else:
        wayo.append(0)
        yo.append(0)
    single.append(int("シングル" in text))
    semi_dou = int("セミダブル" in text) + int("ダブル" in text)
    if semi_dou == 2:
        semidouble.append(1)
        double.append(0)
    elif semi_dou == 1:
        semidouble.append(0)
        double.append(1)
    else:
        semidouble.append(0)
        double.append(0)        
    twin.append(int("ツイン" in text))
    tri.append(int("トリプル" in text))
    omakase.append(int("おまかせ" in text))
    renpaku.append(int("連泊" in text))
add_list = [kinnen,kitsuen,wa,yo,wayo,single,semidouble,double,twin,tri,omakase,renpaku]
add_list = np.array(add_list).T
df_add = pd.DataFrame(add_list,columns = ["禁煙","喫煙","和室","洋室","和洋室","シングル","セミダブル","ダブル","ツイン","トリプル","おまかせ","連泊"])
# print(kinnen)
# print(kituen)
# df_kinnen = pd.DataFrame(kinnen,columns=["禁煙"])
# df_kitsuen = pd.DataFrame(kitsuen,columns=["喫煙"])
# df_wa =  pd.DataFrame(wa,columns=["和室"])
df_s = pd.concat([df_s,df_add], axis=1)
df_s = df_s.drop(columns="text")
df_s
```
